```python

# ==============================================================================
# 1. СБОР И ПОДГОТОВКА ДАННЫХ (Task 1)
# ==============================================================================

# 1.1. Импорт необходимых библиотек
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization

print("TensorFlow Version:", tf.__version__)

# 1.2. Загрузка датасета Malaria из TensorFlow Datasets
(ds_train, ds_info) = tfds.load('malaria', split='train', shuffle_files=True, with_info=True, as_supervised=True)

# 1.3. Определение констант
IMG_SIZE = (128, 128) # Стандартизируем размер изображений
BATCH_SIZE = 32 # Количество изображений для обработки за один шаг

# 1.4. Создание функции для предобработки изображений (A.P2)
def preprocess(image, label):
    image = tf.image.resize(image, IMG_SIZE)
    image = image / 255.0
    return image, label

# 1.5. Разделение данных на обучающую, валидационную и тестовую выборки
DATASET_SIZE = ds_info.splits['train'].num_examples
train_size = int(0.7 * DATASET_SIZE)
val_size = int(0.15 * DATASET_SIZE)
test_size = int(0.15 * DATASET_SIZE)

ds_train = ds_train.shuffle(buffer_size=1000)

train_dataset = ds_train.take(train_size)
remaining_dataset = ds_train.skip(train_size)
val_dataset = remaining_dataset.take(val_size)
test_dataset = remaining_dataset.skip(val_size)

# 1.6. Применение предобработки и создание конвейера данных (tf.data)
train_dataset = train_dataset.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
val_dataset = val_dataset.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
test_dataset = test_dataset.map(preprocess).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

print(f"Обучающая выборка: {len(train_dataset) * BATCH_SIZE} изображений")
print(f"Валидационная выборка: {len(val_dataset) * BATCH_SIZE} изображений")
print(f"Тестовая выборка: {len(test_dataset) * BATCH_SIZE} изображений")

# 1.7. Создание слоя для аугментации данных (A.P2, C.P7)
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.2),
])

train_dataset = train_dataset.map(lambda x, y: (data_augmentation(x, training=True), y),
                                  num_parallel_calls=tf.data.AUTOTUNE)


# ==============================================================================
# 2. РАЗРАБОТКА И ПОСТРОЕНИЕ МОДЕЛИ (Tasks 2 & 3)
# Создание собственной архитектуры CNN с нуля (B.P3)
# ==============================================================================

# 2.1. Построение последовательной модели
model = Sequential()

# Входной слой и первая сверточная группа
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=IMG_SIZE + (3,)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

# Вторая сверточная группа
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

# Третья сверточная группа
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))

# "Выпрямление" данных для полносвязного слоя
model.add(Flatten())

# Полносвязные слои (классификатор)
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5)) # Dropout для борьбы с переобучением (C.P7)

# Выходной слой
# 1 нейрон и sigmoid для бинарной классификации
model.add(Dense(1, activation='sigmoid'))

# 2.2. Компиляция модели
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])

# Вывод структуры построенной модели
model.summary()


# ==============================================================================
# 3. ОБУЧЕНИЕ МОДЕЛИ (Task 3)
# ==============================================================================

# 3.1. Обучение модели (C.P5)
EPOCHS = 25 # Модели с нуля нужно больше времени на обучение
history = model.fit(train_dataset,
                    epochs=EPOCHS,
                    validation_data=val_dataset)


# ==============================================================================
# 4. ОЦЕНКА И НАСТРОЙКА (Task 4)
# ==============================================================================

# 4.1. Визуализация результатов обучения (B.D1)
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(EPOCHS), acc, label='Training Accuracy')
plt.plot(range(EPOCHS), val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(range(EPOCHS), loss, label='Training Loss')
plt.plot(range(EPOCHS), val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()

# 4.2. Оценка модели на тестовой выборке (C.P6)
loss, accuracy, precision, recall = model.evaluate(test_dataset)
print("\n--- Test Set Evaluation ---")
print(f'Test Accuracy: {accuracy:.4f}')
print(f'Test Precision: {precision:.4f}')
print(f'Test Recall: {recall:.4f}')

# 4.3. Построение матрицы ошибок (Confusion Matrix) и отчета о классификации (B.D1)
# --- НАДЕЖНЫЙ МЕТОД ИЗВЛЕЧЕНИЯ МЕТОК И ПРЕДСКАЗАНИЙ ---
y_true = []
y_pred_probs = []

for images, labels in test_dataset:
    preds = model.predict_on_batch(images)
    y_pred_probs.extend(preds.flatten())
    y_true.extend(labels.numpy())

y_true = np.array(y_true)
y_pred_probs = np.array(y_pred_probs)
y_pred = (y_pred_probs > 0.5).astype("int32")

# --- Теперь строим визуализации с правильными данными ---
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Parasitized', 'Uninfected'],
            yticklabels=['Parasitized', 'Uninfected'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

print("\n--- Classification Report ---")
print(classification_report(y_true, y_pred, target_names=['Parasitized', 'Uninfected']))


# ==============================================================================
# 5. РАЗВЕРТЫВАНИЕ (Task 5) - Теоретические шаги
# ==============================================================================

print("\n--- Deployment Considerations ---")
print("1. Model Saving: Save the trained model in a standard format (e.g., HDF5 or SavedModel).")
print("2. Backend Development: Create a server API (e.g., using Flask or FastAPI) to serve the model.")
print("3. Frontend Development: Create a user interface for image upload.")
print("4. Infrastructure: Deploy the system on a cloud or on-premise server.")
print("5. Integration: Integrate with existing hospital systems (PACS, RIS).")
print("6. Monitoring & Maintenance: Continuously monitor and retrain the model.")
```
